# Biblioteka Requests i zapytania HTTP

W ekosystemie Pythona największą popularnością jeśli chodzi o wykonywanie zapytań HTTP cieszy się biblioteka `requests`. Aby z niej skorzystać musimy ją zainstalować standardowo wywołując:
```
pip install requests
```
lub
```
py -m pip install requests
```

Po instalacji możemy ją zaimportować i używać:

````python
import requests

response = requests.get("https://google.com")

print(f"status code: {response.status_code}, body:\n {response.text}")
````

Każdy z czasowników HTTP ma odpowiadającą funkcję w module requests o tej samej nazwie pisanej małymi literami - np. `POST` to `requests.post`, `DELETE` to `requests.delete` etc.

Wynikiem działania tych funkcji są obiekty `Response`, które zawierają przydatne pola i metody, pozwalające na dostanie się do odpowiedzi HTTP zwróconej w wyniku zapytania. W przykładzie powyżej pole `status_code` przechowuje kod odpowiedzi HTTP, a `text` skonwertowane na `str` w formatowaniu `UTF-8` ciało (body) odpowiedzi.

Jeśli body odpowiedzi jest w formacie `JSON` można użyć metody `.json()` by dostać słownik:

````python
import requests

response = requests.get("https://dummyjson.com/products/1")

for k, v in response.json().items():
    if isinstance(v, list):
        print("[")
        for x in v:
            print(f"   {x}")
        print("]")
        continue
    print(f"klucz: {k} -> wartość: {v}")
````

Aby zobaczyć nagłówki odpowiedzi, można użyć pola `.headers`, które jest (prawie) słownikiem:

````python
import requests

response = requests.get("https://dummyjson.com/products/1")
for k, v in response.headers.items():
    print(f"{k}: {v}")
````

Prawie, ponieważ dobierając się do wartości nagłówków po ich nazwie nie musimy przejmować się wielkością liter:

````python
import requests

response = requests.get("https://dummyjson.com/products/1")
print(response.headers["Content-Length"])
print(response.headers["content-length"])
````

### Przesyłanie danych w ciele zapytania

Często chcemy wysłać dane w ciele zapytania np. `POST`. Aby to osiągnąć możeemy użyć parametru `data` odpowiedniej funkcji:

````python
import requests

response = requests.post("https://dummyjson.com/products/add", data={
    'title': 'BMW Pencil',
  })
print(response.status_code)
````

Tak wysłane dane są kodowane jako `application/x-www-form-urlencoded`. Jeśli chcemy je przesłać jako JSON, możemy użyć parametru `json`:

````python
import requests

response = requests.post("https://dummyjson.com/products/add", json={
    'title': 'BMW Pencil',
  })
print(response.status_code)
````

Biblioteka Requests automatycznie ustawi odpowiedni `Content-Type`, natomiast jeśli chcemy nadpisać któryś z nagłówków zapytania możemy użyć parametru `headers`:

````python
import requests

response = requests.post("https://dummyjson.com/products/add", json={
    'title': 'BMW Pencil',
  },
  headers={'Content-Type': 'application/json'}
)
print(response.status_code)
````

### Timeouty

Możemy wymusić timeout zapytania po zadanym czasie:

````python
import requests
from requests.exceptions import Timeout

try:
    response = requests.get('https://dummyjson.com/products', timeout=1)
except Timeout:
    print('Timeout occured!')
else:
    print('Request succeeded!')
````

Timeout jest w tym wypadku podawany w sekundach i może przyjmować wartości zmiennoprzecinkowe.

### Obiekty Session
Pod spodem biblioteka `requests` używa specjalnej klasy `Session`, przechowującej różne parametry połączeń i zapytań HTTP między zapytaniami. Można z niej skorzystać, gdy chcemy dostroić wydajność w programach, gdzie zapytań robi się dużo lub po prostu ustawić np. wspólne wymagane nagłówki lub inne parametry połączeń:

````python
import requests
from requests.exceptions import Timeout

with requests.Session() as session:
    session.auth = ("username", "password")
    response = requests.get('https://dummyjson.com/products')
````
