# Czym jest Python i do czego można go zastosować?

Python to dynamicznie typowany, interpretowany język ogólnego przeznaczenia. Robi się w nim wszystko, bo do większości zastosowań jest po prostu wystarczająco dobry. Jego bolączką bywa wydajność i trudna relacja z współbieżnością, ale zazwyczaj albo to nie jest problem albo jest już moduł napisany w C, który to obchodzi.

## Język kompilowany czy interpretowany?
Mimo że Python jest uważany za język interpretowany, w rzeczywistości sprawa jest bardziej skomplikowana. Po pierwsze istnieje wiele implementacji języków programowania, w tym Pythona (CPython jest "kanoniczną", ale jest też wiele innych np. pypy, Cython, IronPython etc.), więc nie można powiedzieć o języku, że jest "interpretowany".

Implementacje języków programowania dzielą się na dwie główne klasy: kompilowane i interpretowane. Implementacje kompilowane wymagają budowania, czyli tłumaczenia kodu napisanego przez programistę na kod maszynowy, wykonywalny przez komputer. Bez tego kod jest dla komputera kompletnie niezrozumiały. Plusem jest natomiast możliwość optymalizacji kodu przed wykonaniem i możliwość wyprodukowania samowystarczalnego pliku wykonywalnego.

Języki interpretowane umożliwiają pisanie programów, które można od razu uruchomić - ich kod jest - nomen omen -interpretowany, lub mówiąc prościej wykonywany przez interpreter. W najprostszej formie kod napisany przez programistę jest wprost "rozumiany" przez program interpretera, jednak ze względów wydajnościowych jest to rozwiązanie niepraktyczne i rzadko stosowane. Zazwyczaj kod jest przerabiany na prostszą reprezentację pośrednią (IR), optymalizowany, a interpreter wykonuje już wstępnie przetworzony kod. W przypadku najpopularniejszej, "kanonicznej" implementacji Pythona mamy do czynienia właśnie z taką sytuacją - polecenie `python` w rzeczywistości kompiluje kod do tzw. bytecode'u, cache'owanego w postaci plików `.pyc`, a następnie szybka maszyna wirtualna wykonuje bytecode. Dzięki temu zyskujemy przenośność kodu charakterystyczną dla implementacji interpretowanych oraz akceptowalną do większości zastosowań szybkość. Ceną jest konieczność zainstalowania interpretera Pythona - jeśli jest to problem, warto rozważyć implementacje takie jak Cython, które kompilują podzbiór Pythona do kodu maszynowego.

## Python 2.x vs. Python 3.x
Przez wiele lat kiedy język znacząco zyskał na popularności, Python wystepował w wersjach 2.x i do tej pory niejedna aplikacja, z którą można się spotkać jest kompatybilna z własnie tymi wersjami. Proces przejścia na znacząco ulepszoną, ale niekompatybilną wstecznie wersję 3 zajął ponad 12 lat, a w tym czasie wiele ulepszeń z wersji 3.x zostało przeniesionych do wersji 2.7.x, co zniechęciło wielu developerów do przenosin. Obecnie jednak wszystkie wersje 2.x sa niewspierane i nie powstaje właściwie nowy kod kompatybilny z tą wersją. Co więcej, wszystkie powszechnie używane biblioteki i narzędzia są już kompatybilne z wersją 3.x. Z tego powodu podczas tych zajęć nie będziemy poświęcać czasu na analizowanie różnic między nimi - zainteresowanych zachęcam do przeczytania porównania [tutaj](https://www.pythonforbeginners.com/development/python-2-vs-python-3-examples), a wykorzystywać będziemy najnowszą wersję 3.9.

## Korzystanie z Jupytera
Jupyter umożliwia tworzenie notebooków, takich jak ten - łączących markdown z wykonywalnym kodem Pythona. Z punktu widzenia użytkownika przydatne jest wiedzieć, że Jupyter jest podzielony na "komórki", które mogą być osobno ewaluowane i zmieniane, oraz że ma 2 tryby:
- command - w tym trybie możliwe jest ewaluowanie komórek, przesuwanie komórek, wycinanie, wklejanie, zmienianie typu itp.
- edit - w tym trybie można edytować zawartość komórek

### Przydatne skróty klawiszowe:

| Skrót | Akcja |
| :--- | :--- |
|`Enter` | włącza tryb edycji zaznaczonej komórki |
|`Esc` | włącza tryb command |
|`Ctrl + Enter` | ewaluuje komórkę |
|`Alt + Enter` | ewaluuje komórkę i tworzy nową poniżej |
|`Shift + Enter` | ewaluuje komórk i zaznacza poniżej |
| m | zmienia typ komórki na markdown |
| y | zmienia typ komórki na kod |
| d+d | usuwa komórkę |
| o | chowa/pokazuje wynik ewaluacji kodu |
| a | tworzy komórkę powyżej |
| b | tworzy nową komórkę poniżej |
